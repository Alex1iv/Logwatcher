````python
import os 
import sys
#set path
sys.path.insert(1, '../')
from datetime import datetime
import re
````

```sql
CREATE TABLE log_events (
    id SERIAL PRIMARY KEY,
    timestamp TIMESTAMP NOT NULL,
    switch_name TEXT,
    error_type TEXT,
    mac TEXT,
    vlan INTEGER,
    src TEXT,
    dst TEXT,
    raw TEXT,
    created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
);

CREATE INDEX idx_log_ts ON log_events(timestamp);
CREATE INDEX idx_log_mac ON log_events(mac);
CREATE INDEX idx_log_err ON log_events(error_type);
```

````python
LOG_REGEX = re.compile(
    r'^(?P<timestamp>.{20})\s+(?P<switch_name>\S+)\s+(?P<err_type>%\S+):\s+'
    r'Mac\s+(?P<mac>\S+)\s+in\s+vlan\s+(?P<vlan>\d+)\s+has\s+moved\s+from\s+'
    r'(?P<src>\S+)\s+to\s+(?P<dst>\S+)'
)
pattern = r"^(?P<timestamp>[\d]{4} \w{3} [\d]{1,2} [\d]{2}:[\d]{2}:[\d]{2}) (?P<host>[\w\._-]+) %"
pattern += r"(?P<facility>\w+)-(?P<severity>\d)-(?P<mnemonic>\w+): (?P<message>.*)$"

MOVE_REGEX = re.compile(
    r'^(?P<timestamp>\d{4}\s+\w{3}\s+\s*\d+\s+\d{2}:\d{2}:\d{2})\s+'
    r'(?P<switch_name>\S+)\s+(?P<error_type>%\S+):\s+'
    r'Mac\s+(?P<mac>[0-9a-fA-F\.]+)\s+in\s+vlan\s+(?P<vlan>\d+)\s+has\s+moved\s+from\s+'
    r'(?P<src>\S+)\s+to\s+(?P<dst>\S+)',
    re.IGNORECASE
)
````
